#주어진 대여소번호와 대여일시에서 정기권이용자와 전체이용자 계산하기

```python
data = pd.read_csv('/content/drive/MyDrive/데이터셋/0_raw_data/raw_data/서울특별시 공공자전거 이용정보(시간대별)_22.05.csv', encoding = 'cp949')
data = data[['대여일자', '대여시간', '대여소번호', '대여구분코드']]
data['대여일시'] = pd.to_datetime(list(data['대여일자'].astype('str') + ' ' + data['대여시간'].astype('str') + ':00' +':00'))

data_1 = data[data['대여구분코드'] == '정기권']
data_1 = data_1[['대여소번호', '대여일시']]
data_1 = pd.DataFrame(data_1.groupby(['대여소번호', '대여일시']).value_counts()).reset_index()
data_1.columns = ['대여소번호', '대여일시', '정기권이용자수']

data = data[['대여소번호', '대여일시']]
data = pd.DataFrame(data.groupby(['대여소번호', '대여일시']).value_counts()).reset_index()
data.columns = ['대여소번호', '대여일시', '전체이용자수']

data_1 = pd.merge(data, data_1, how = 'left', left_on = ['대여소번호', '대여일시'], right_on = ['대여소번호', '대여일시'])
data_1 = data_1.fillna(0)
data_1 = data_1[['대여소번호', '대여일시', '정기권이용자수', '전체이용자수']]
data_1
```
