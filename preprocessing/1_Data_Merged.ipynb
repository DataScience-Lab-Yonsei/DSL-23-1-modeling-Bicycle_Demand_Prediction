#겹치는 대여소만 남기기
1. 대여 대여소와 반납대여소 수가 다른 경우, 더 적은 것을 택함
2. 모든 월의 대여소 목록 중 겹치는 것만 남기기

```python
list_all = list(set(data_1['대여 대여소번호'].unique()) & set(data_1['반납대여소번호'].unique()))

for n, i in enumerate([data_1, data_2, data_3, data_4, data_5, data_6]):
  list_i = list(set(i['대여 대여소번호'].unique()) & set(i['반납대여소번호'].unique())) #주어진 월 데이터에서 대여 대여소와 반납 대여소가 겹치는 대여소만 추출
  list_temp = list(set(list_all) & set(list_i)) #총 대여소 리스트와 신규 대여소 리스트 중 겹치는 부분만 남기기
  list_all = list_temp #총 대여소 리스트 업데이트
  print('{}번째 데이터프레임의 대여소의 수 : '.format(n+1), len(list_i))
  print('{}번째 대여소 리스트까지 반영한 결과 겹치는 대여소의 수 : '.format(n+1), len(list_all))

```
